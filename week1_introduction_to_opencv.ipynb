```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2 as cv

# mengambil gambar dan menyimpan nilai matrixnya ke dalam variabel 'img'
img = cv.imread('/content/drive/MyDrive/Colab Notebooks/pemrosesan-citra/daun.jpeg')

# menampilkan img
cv2_imshow(img)

# img[y, x]

# berisi y * x pixel serta nilai dari matrix tersebut (rgb? binary?)
size_of_img = img.shape

# skala, variabel yang digunakan untuk mengambil besaran luas pixel
# berguna untuk menentukan besar pixel yang di-cut
scale = 10

# mengambil tinggi dan lebar gambar
height = size_of_img[0]
width = size_of_img[1]

# mengambil titik tengah dari gambar
centerY = int(height/2 - scale/2)
centerX = int(width/2 - scale/2)

# variabel yang berisi gambar pattern
## untuk variabel slices, nilai dari variabel tersebut mengambil titik tengah
## dengan besaran berdasarkan nilai dari variabel 'scale'
slices = img[centerY:centerY+scale, centerX:centerX+scale]
black = np.random.randint(1, size=(10, 10, 1))

# menampilkan isi dari variabel 'size_of_img'
print(size_of_img)

# digunakan sebagai variabel pembantu untuk melakukan 
# looping untuk implementasi pattern
counterPattern = 0
pattern = 1357

# looping tiap matrix dari image, digunakan if else untuk memberi
# jarak antar pattern di image
for x in range(len(img[0])):
  for y in range(len(img)):
    if(counterPattern < pattern):
      counterPattern += 1
    else:
      counterPattern = 0
      if(x+scale < width and y+scale < height):
        img[y:y+scale, x:x+scale] = slices

# menampilkan gambar yang sudah diberi pattern
cv2_imshow(img)
```
