## Get Array

```python
# 3x3
# getting an array based on the coordinate

def getArrayFull(y,x,img):
  array = [img[y+1,x-1],
           img[y+1,x],
           img[y+1,x+1],

           img[y,x-1],
          #  titik tengah y,x
           img[y,x+1],

           img[y-1,x-1],
           img[y-1,x],
           img[y-1,x+1]
           ]
  return array

def getArrayBot(y,x,img):
  array = [
          #  img[y+1,x-1],
          #  img[y+1,x],
          #  img[y+1,x+1],

           img[y,x-1],
          #  titik tengah y,x
           img[y,x+1],
           
           img[y-1,x-1],
           img[y-1,x],
           img[y-1,x+1]
           ]
  return array

def getArrayTop(y,x,img):
  array = [
           img[y+1,x-1],
           img[y+1,x],
           img[y+1,x+1],
      
           img[y,x-1],
          #  titik tengah y,x
           img[y,x+1],

          #  img[y-1,x-1],
          #  img[y-1,x],
          #  img[y-1,x+1]
           ]
  return array

def getArrayTopLeft(y,x,img):
  array = [
          #  img[y+1,x-1],
           img[y+1,x],
           img[y+1,x+1],

          #  img[y,x-1],
          #  titik tengah y,x
           img[y,x+1],

          #  img[y-1,x-1],
          #  img[y-1,x],
          #  img[y-1,x+1]
           ]
  return array

def getArrayTopRight(y,x,img):
  array = [
           img[y+1,x-1],
           img[y+1,x],
          #  img[y+1,x+1],

           img[y,x-1],
          #  titik tengah y,x
          #  img[y,x+1],

          #  img[y-1,x-1],
          #  img[y-1,x],
          #  img[y-1,x+1]
           ]
  return array

def getArrayBotLeft(y,x,img):
  array = [
          #  img[y+1,x-1],
          #  img[y+1,x],
          #  img[y+1,x+1],

          #  img[y,x-1],
          #  titik tengah y,x
           img[y,x+1],

          #  img[y-1,x-1],
           img[y-1,x],
           img[y-1,x+1]
           ]
  return array

def getArrayBotRight(y,x,img):
  array = [
          #  img[y+1,x-1],
          #  img[y+1,x],
          #  img[y+1,x+1],

           img[y,x-1],
          #  titik tengah y,x
          #  img[y,x+1],

           img[y-1,x-1],
           img[y-1,x],
          #  img[y-1,x+1]
           ]
  return array

def getArrayLeft(y,x,img):
  array = [
          #  img[y+1,x-1],
           img[y+1,x],
           img[y+1,x+1],

          #  img[y,x-1],
          #  titik tengah y,x
           img[y,x+1],

          #  img[y-1,x-1],
           img[y-1,x],
           img[y-1,x+1]
           ]
  return array

def getArrayRight(y,x,img):
  array = [
           img[y+1,x-1],
           img[y+1,x],
          #  img[y+1,x+1],

           img[y,x-1],
          #  titik tengah y,x
          #  img[y,x+1],

           img[y-1,x-1],
           img[y-1,x],
          #  img[y-1,x+1]
           ]
  return array
```

## Minmax

```python
# minmax algorithm

def minmax_help(y, x, img):
  if(y == 0): #pojok atas
    if(x == 0): #atas kiri
      array = getArrayTopLeft(y,x,img)
    elif(x == len(img[0])-1): #atas kanan
      array = getArrayTopRight(y,x,img)
    else: 
      array = getArrayTop(y,x,img)

  elif(y == len(img)-1):  #pojok bawah
    if(x == 0): #bawah kiri
      array = getArrayBotLeft(y,x,img)
    elif(x == len(img[0])-1): #bawah kanan
      array = getArrayBotRight(y,x,img)
    else:
      array = getArrayBot(y,x,img)

  elif(x == 0):
    array = getArrayLeft(y,x,img)

  elif(x == len(img[0])-1):
    array = getArrayRight(y,x,img)

  else:
    array = getArrayFull(y,x,img)

  low = min(array)
  high = max(array)

  # print(array)

  tengah = img[y,x]

  if(tengah < low):
    return low
  elif(tengah > high):
    return high
  else:
    return tengah

def minmax(img):
  for y in range(len(img)):
    for x in range(len(img[0])):
      img[y,x] = minmax_help(y,x, img)
  return img
```

## Average

```python
def mean(img):
  for y in range(len(img)):
    for x in range(len(img[0])):
      img[y,x] = mean_help(y,x,img)
  return img

def mean_help(y,x,img):
  if(y == 0): #pojok atas
    if(x == 0): #atas kiri
      array = getArrayTopLeft(y,x,img)
    elif(x == len(img[0])-1): #atas kanan
      array = getArrayTopRight(y,x,img)
    else: 
      array = getArrayTop(y,x,img)

  elif(y == len(img)-1):  #pojok bawah
    if(x == 0): #bawah kiri
      array = getArrayBotLeft(y,x,img)
    elif(x == len(img[0])-1): #bawah kanan
      array = getArrayBotRight(y,x,img)
    else:
      array = getArrayBot(y,x,img)

  elif(x == 0):
    array = getArrayLeft(y,x,img)

  elif(x == len(img[0])-1):
    array = getArrayRight(y,x,img)

  else:
    array = getArrayFull(y,x,img)

  array.append(img[y,x])

  avr = average(array)
  # print(array)
  return avr

def average(array):
  return sum(array)/len(array)
```

## Median

```python
import statistics

def median(img):
  for y in range(len(img)):
    for x in range(len(img[0])):
      img[y,x] = median_help(y,x,img)
  return img

def median_help(y,x,img):
  if(y == 0): #pojok atas
    if(x == 0): #atas kiri
      array = getArrayTopLeft(y,x,img)
    elif(x == len(img[0])-1): #atas kanan
      array = getArrayTopRight(y,x,img)
    else: 
      array = getArrayTop(y,x,img)

  elif(y == len(img)-1):  #pojok bawah
    if(x == 0): #bawah kiri
      array = getArrayBotLeft(y,x,img)
    elif(x == len(img[0])-1): #bawah kanan
      array = getArrayBotRight(y,x,img)
    else:
      array = getArrayBot(y,x,img)

  elif(x == 0):
    array = getArrayLeft(y,x,img)

  elif(x == len(img[0])-1):
    array = getArrayRight(y,x,img)

  else:
    array = getArrayFull(y,x,img)

  array.append(img[y,x])

  mid = statistics.median(array)
  # print(array)
  return mid
```
